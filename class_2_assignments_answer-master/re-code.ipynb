```python
frequneces_all=[f for w,f in word_counts.most_common()]  
sum_of_frequences=sum(frequneces_all)   #求出词频的总和, 这样是不是直接就是等于len(valida_TEXT)?
sum_of_frequences
```

```python
eps=1/sum_of_frequences    #eps表示epsilon 防止要预测的单词早语料库中没有，为了防止在连乘中出现一个0，所以给这样的单词的概率就是eps.
eps
```

```python


def get_prob(tokens):
    if tokens in word_counts:
        tokens_frequence=word_counts[tokens]/sum_of_frequences
        return tokens_frequence
    else: return eps
    
get_prob("我们")
    
```

```python
def product(numbers):                 #定义一个连乘的函数
    return reduce(lambda n1,n2:n1*n2,numbers)   #refuce是啥意思？
product([1,2,3])    #里面是个list


def language_model_one_gram(string):   #定义one_gram的语言模型
    word = cut(string)
    return product([get_prob(w) for w in word])

language_model_one_gram("休斯顿火箭队取得五连胜")
```

## 2-gram

$$ Pr(S) = Pr(w_1\cdot w_2 \cdots w_n) \sim Pr(w_1) \cdot Pr(w_2 | w_1) \cdots Pr(w_n | w_{n-1})$$

$$ Pr(w_2 | w_1) =  \frac { Pr(w_1 \cdot w_2) }{Pr(w_1)} $$ 

```python

# from icecream import ic

all_2_grams_words=["".join(ALL_TOKENS[i:i+2]) for i in range(len(ALL_TOKENS[:-2]))]
_2_grams_words_sum=len(all_2_grams_words)
_2_grams_words_counter=Counter(all_2_grams_words)

def get_combine_prop(w1,w2):
    if w1+w2 in ALL_TOKENS: return _2_grams_words_counter[w1+w2]/_2_grams_words_sum
    else:
        return 1/_2_grams_words_sum
    
def get_2_gram_prop(w1,w2):
    return get_combine_prop(w1,w2)/get_prob(w1)

def language_model_of_2_gram(sentence):
    sentence_propability=1
    words=cut(sentence)
    
    for i, word in enumerate(words):
        if i == 0:
            prop=get_prob(word)
        else:
            previous=words[i-1]
            prop=get_2_gram_prop(previous,words[i])   
            sentence_propability*=prop #这里为什么没用product函数
    return sentence_propability  
```

```python
language_model_of_2_gram("休斯顿火箭队排名西部第二")
```
