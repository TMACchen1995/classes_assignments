## Dynamic Progarmming problem（不断查表的意思）
#### 1. 知道重复的子问题
#### 2. 将重复的子问题，写个装饰器存储起来
#### 3. solution要进行解析

```python
price_list=[1,5,8,9,10,17,17,20,24,30]

from collections import defaultdict
price=defaultdict(int)

for i, p in enumerate(price_list):
    price[i+1] = p
    
assert(price[1])==1


solution = {}

def r(n):
    max_price, split_point=max(
        
        [(price[n],0)] + [(r(i)+r(n-i),i) for i in range(1,n)], key=lambda x: x[0]
        
        
    )
    solution[n] = (max_price,split_point)
    
    return max_price

r(5)
    
solution

#The weakmness of the produce is that when the arg is relatively large, the used time can be very long, because of the overlapping part
```

## 子问题进行存储

```python
#Writing a decorator to enhance the efficient
import time

def memo(func):
    cache={}
#     @wraps(func)
    def _wrap(arg):
#         arg = kwargs.values()
#         print(arg)
        if arg in cache: res = cache[arg]
        else:
            res = func(arg)
            cache[arg] = res
        return res
    return _wrap


@memo
def r(n):
    max_price, split_point=max(
        
        [(price[n],0)] + [(r(i)+r(n-i),i) for i in range(1,n)], key=lambda x: x[0]
        
        
    )
    solution[n] = (split_point, n - split_point)
    
    return max_price
    

r(333)


#
```

```python
#Another way: using decorator @lru_cache which can be import from functools. 

from functools import lru_cache

@lru_cache(maxsize=2**10)
def r(n):
    max_price, split_point=max(
        
        [(price[n],0)] + [(r(i)+r(n-i),i) for i in range(1,n)], key=lambda x: x[0]
        
        
    )
    solution[n] = (split_point, n - split_point)
    
    return max_price
    

r(333)
```

## 对solution进行解析

```python
#Show how the target_length is parsed in a clear way.   递归思想！！！！
def not_cut(split): return split == 0

def parse_solution(target_length, solution):
    left, right = solution[target_length]
    if not_cut(left): return [right]      #仔细观察该递归的出口。当solution[para] 返回的tuple的第一个元素为零时，则该para就不再分割即为出口。
    
    return parse_solution(left,solution) + parse_solution(right, solution)   #左右两边均不为零时， 则对左右两边同时解析。

parse_solution(67,solution)
```

## Edit Distance

```python

#Edit Distance is similar with the example above, the idea also is dynamic programming.
#The first step: find the overlapping part , in this situation, there are three all transformatio rule together, 
#until the string1 == string2, then get the min 
@memo
def edit_distance(string1,string2):
    if len(string1) == 0: return(len(string2))
    if len(string2) == 0: return (len(string1))
    
    tail_s1 = string1[-1]
    tail_s2 = string2[-1]
    
    min_edit_distance=min([
        
        edit_distance(string1[:-1], string2) + 1,      #situation 1, string1 del tail
        edit_distance(string1, string2[:-1]) + 1,      #situation 2, string1 add the tailof string2
        edit_distance(string1[:-1], string2[:-1]) + (0 if tail_s1 ==  tail_s2 else 2)    #situation 3
         
    ])
    
    return min_edit_distance


edit_distance("beijing", "beij")
```
